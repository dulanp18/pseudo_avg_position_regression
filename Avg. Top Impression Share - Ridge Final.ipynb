```python
df_final=df[['top_is','abs_top_is','avg_pos']].drop_duplicates()
Y = df_final['avg_pos']
X = df_final[['top_is','abs_top_is']]
X['top_is'] = X['top_is']
X['abs_top_is'] = X['abs_top_is']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Adjust alpha based on previous result
alpha=np.arange(1, 11.6, 0.8)
rmse=[]
r_2=[]

for alph in alpha:
    ridge=Ridge(alpha=alph, copy_X=True, fit_intercept=True)
    ridge.fit(X_train, y_train)
    predict=ridge.predict(X_test)
    rmse.append(np.sqrt(mean_squared_error(predict, y_test)))
    r_2.append(ridge.score(X_test,y_test))
    
print(r_2)
plt.scatter(alpha, r_2)

```
